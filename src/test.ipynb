```python
from functions.utils import *

file = load_file(r'D:\GithubRepositories\RAG-system\docs\doc.txt')
print(file)
```

```python
from langchain_text_splitters import RecursiveCharacterTextSplitter

splitter = RecursiveCharacterTextSplitter(chunk_size=1000, chunk_overlap=200)
chunks = splitter.split_documents(file)
for chunk in chunks:
    print(chunk)
    print("-" * 30)
```

```python
import json

json_docs = [
    {"page_content": d.page_content, "metadata": d.metadata} for d in chunks
]

print(json_docs)
```

```python
with open('docs.json', 'w', encoding='utf8') as f:
    json.dump(json_docs, f, ensure_ascii=False, indent=2)
```

```python
import json

for chunk in chunks:
    x = json.dumps({"metadata" : chunk.metadata,"page_content" : chunk.page_content}, ensure_ascii=False, indent=4)
    print(x)
```

```python
docs = []

with open('docs.json', 'r', encoding='utf8') as f:
    docs = json.load(f)

print(docs)
```

```python
new_docs = [{'page_content' : "doc1", "metadata":{'source' : "doc1.dc"}}, {'page_content' : "doc2", "metadata":{'source' : "doc2.dc"}}]
docs.extend(new_docs)
print(docs)
```

```python
from langchain_core.documents import Document

documents = []
for doc in docs:
    documents.append(Document(page_content = doc["page_content"], metadata = doc["metadata"]))
```
